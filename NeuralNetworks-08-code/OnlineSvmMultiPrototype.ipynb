### Online Multiclass Multiprototype Linear SVM

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.io as sio
from tqdm import tqdm
import matplotlib.animation as animation
```

```python
%matplotlib inline
mpl.rcParams['figure.dpi']= 150
```

```python
MnistTrainX = sio.loadmat ('../../datasets/mnist/MnistTrainX')['MnistTrainX'] / 255;
MnistTrainY = sio.loadmat ('../../datasets/mnist/MnistTrainY')['MnistTrainY'];
MnistTestX  = sio.loadmat ('../../datasets/mnist/MnistTestX')['MnistTestX'] / 255;
MnistTestY  = sio.loadmat ('../../datasets/mnist/MnistTestY')['MnistTestY'];
```

```python
XTrain = MnistTrainX
yTrain = MnistTrainY.squeeze()
```

```python
margin = 1
learning_rate = 0.01
weight_decay = 0.00001
K = 7
C = 10
maxEpoch = 20
(N, dim) = XTrain.shape
W = np.random.rand(K*C, dim) / 784
b = np.zeros ([K*C])

fig = plt.figure()
metadata = dict(title='OnlineSVM7Prototype', 
                artist='Matplotlib',
                comment='Movie support!')
pillowWriter = animation.writers['pillow']
moviewriter = pillowWriter(fps=4, metadata=metadata)
moviewriter.setup(fig=fig, outfile='OnlineSVM7Prototype.gif', dpi=150)

for i in tqdm (range(maxEpoch)):
    for j in range (N):
        x = XTrain[j,:]
        t = yTrain[j]
        z = (W @ x) + b 
        
        p = t*K + np.argmax(z[t*K:(t+1)*K])
        zz = z.copy()
        zz[t*K:(t+1)*K] = -np.inf
        q = np.argmax (zz)
        W *= 1 - learning_rate * weight_decay
        b *= 1 - learning_rate * weight_decay
        if (z[p] - z[q] < margin):
            #print (z[p] - z[q])
            W[p,:] += learning_rate * x            
            W[q,:] -= learning_rate * x
            b[p]   += learning_rate * 1
            b[q]   -= learning_rate * 1
        if (j % 10000==9999):
            #print (z[p] - z[q])
            wholeImage = np.zeros ([K * 29, C * 29, 3])
            for u in range (C):
                for v in range (K):
                    img = np.reshape(W[u*K+v,:], [28,28])
                    img = img / np.max(abs(img))
                    posImage = np.maximum(0 , img)
                    negImage = np.maximum(0 , -img)
                    wholeImage[v * 29:(v+1)* 29 -1, u * 29:(u+1) * 29-1, 0] = posImage
                    wholeImage[v * 29:(v+1)* 29 -1, u * 29:(u+1) * 29-1, 1] = negImage
            plt.imshow (wholeImage)
            plt.axis('off')
            moviewriter.grab_frame()
moviewriter.finish()
```

```python
XTest = MnistTestX
yTest = MnistTestY

score = 0.0
for j in range (XTest.shape[0]):
    x = XTest[j,:]
    t = yTest[j]
    value = W @ x + b
    max_idx = np.argmax(value)
    y = max_idx // K
    if (y == t):
        score += 1
print ('Accuracy on test data = {0}\%'.format (score/XTest.shape[0] * 100))
```
