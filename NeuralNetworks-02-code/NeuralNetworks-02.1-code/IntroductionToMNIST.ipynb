# Introduction to the MNIST dataset

```python
import scipy.io as sio
from skimage import data
from skimage.transform import resize
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
MnistTrainX = sio.loadmat ('../../datasets/mnist/MnistTrainX')['MnistTrainX'];
MnistTrainY = sio.loadmat ('../../datasets/mnist/MnistTrainY')['MnistTrainY'];
MnistTestX  = sio.loadmat ('../../datasets/mnist/MnistTestX')['MnistTestX'];
MnistTestY  = sio.loadmat ('../../datasets/mnist/MnistTestY')['MnistTestY'];
```

```python
%matplotlib inline
mpl.rcParams['figure.dpi']= 600
```

```python
Rows = 15
Cols = 15
img = np.zeros([Rows*29,Cols*29,3])

for r in range(Rows): # 0 ... Rows-1
    for c in range(Cols): # 0 ... Cols-1
        imgNo = c*Rows+r
        singleImage = np.reshape(MnistTrainX[imgNo,:],[28,28]) / 255.0
        img[r*29:(r+1)*29-1,c*29:(c+1)*29-1,0] = singleImage
        img[r*29:(r+1)*29-1,c*29:(c+1)*29-1,1] = singleImage
        img[r*29:(r+1)*29-1,c*29:(c+1)*29-1,2] = singleImage

plt.imshow (img, cmap='gray');
plt.axis('off')
plt.tight_layout(pad=0)
plt.imsave('mnistsamples.png', img)
plt.show()
```

```python
Rows = 15
Cols = 15
img = np.zeros([Rows*29,Cols*29,3])

imgNo = -1
for r in range(Rows): # 0 ... Rows-1
    for c in range(Cols): # 0 ... Cols-1
        imgNo += 1
        while (MnistTrainY[imgNo]!= 2):
            imgNo+=1
        singleImage = np.reshape(MnistTrainX[imgNo,:],[28,28]) / 255.0
        img[r*29:(r+1)*29-1,c*29:(c+1)*29-1,0] = singleImage
        img[r*29:(r+1)*29-1,c*29:(c+1)*29-1,1] = singleImage
        img[r*29:(r+1)*29-1,c*29:(c+1)*29-1,2] = singleImage

plt.imshow (img, cmap='gray');
plt.axis('off')
plt.tight_layout(pad=0)
plt.imsave('mnist_samples_of_2.png', img)
plt.show()
```
